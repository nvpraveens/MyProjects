```python
X
```

```python
#filling na as 0 for now,maybe replace with mean
X=X.fillna(0)
y=y.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3, random_state=0)
```

### Build multiple Machine learning models for the above set

#### Modelling Decision Tress

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
```

```python
model = DecisionTreeClassifier()
model.fit(X_train,y_train)
```

```python
predictions=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,predictions)
```

```python
predict_2020=model.predict(X)
accuracy_score(y,predict_2020)
```

### KNN with varying K

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
accuracy_rate = []
for i in range(1,100):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    acc = accuracy_score(y_test, pred_i)
    accuracy_rate.append(acc)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,100),accuracy_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Accuracy Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=40)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
accuracy_score(y_test, pred)
```

### Polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Polynomial Regression
for i in range (1,3):
    polynomial_features= PolynomialFeatures(degree=i)

    # transforming x
    x_poly = polynomial_features.fit_transform(X)

    # instantiating the model using linear regression
    model = LinearRegression()

    # fitting the quadratic model
    model.fit(x_poly, y)

    # prediction
    y_poly_pred = model.predict(x_poly)

    # performance evaluation
    print(i)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_poly_pred))  
    print('Mean Squared Error:', metrics.mean_squared_error(y, y_poly_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_poly_pred)))
    print('R2_score:',metrics.r2_square(y,y_poly_pred))
```
