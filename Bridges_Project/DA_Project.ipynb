```python
abs(Master.corr()["Det_Rate"][abs(Master.corr()["Det_Rate"])>0.5].drop('Det_Rate')).index.tolist
```

```python
corrmat = Master.corr()
corrdf = corrmat.where(np.triu(np.ones(corrmat.shape), k=1).astype(bool))
corrdf = corrdf.unstack().reset_index()
corrdf.columns = ['var1', 'var2', 'Correlation']
corrdf.dropna(subset = ['Correlation'], inplace = True)
corrdf['Correlation'] = round(corrdf['Correlation'], 2)
matrix= corrdf.sort_values(by = 'Correlation', ascending = False).head(30)
print("Correlation Coefficient (descending order):\n", matrix)
```
