### KNN with varying K

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
accuracy_rate = []
for i in range(1,100,10):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    acc = accuracy_score(y_test,pred_i)
    accuracy_rate.append(acc)
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,100,10),accuracy_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Accuracy Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy Rate')
```

```python
knn = KNeighborsClassifier(n_neighbors=40)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)
accuracy_score(y_test, pred)
```

### Polynomial regression

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
#Polynomial Regression
for i in range (1,4):
    polynomial_features= PolynomialFeatures(degree=i)

    # transforming x
    x_poly = polynomial_features.fit_transform(X)

    # instantiating the model using linear regression
    model2 = LinearRegression()

    # fitting the quadratic model
    model2.fit(x_poly, y)

    # prediction
    y_poly_pred = model2.predict(x_poly)

    # performance evaluation
    print('For degree:',i)
    print('Mean Absolute Error:', metrics.mean_absolute_error(y, y_poly_pred))  
    print('Mean Squared Error:', metrics.mean_squared_error(y, y_poly_pred))  
    print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y, y_poly_pred)))
    print('R2_score:',metrics.r2_score(y,y_poly_pred))
```

```python
def stackedBar(Freq):
    Freq=Freq.div(Freq.sum())
    Freq=Freq.transpose()
    Freq
    bottom_list = []

    # Loops through each rating column. 
    # Each column will take the sum of all columns before it as its base.
    for i in Freq.columns:
        b = list(Freq.loc[:, Freq.columns[0]: i].sum(axis=1))
        bottom_list.append(b)

    fig, ax = plt.subplots(figsize=(7,7))

    x_ax = Freq.index

    for i, v in enumerate(Freq.columns):
        if i == 0:
            # First column is at the bottom of the chart so bottom is 0 and doesn't need to be specified.
            plt.bar(x_ax, Freq[v],width=1)
        else:
            plt.bar(x_ax, Freq[v], bottom=bottom_list[i - 1],width=0.5)
    # y-axis formatting
    y_vals = ax.get_yticks()
    ax.set_yticklabels(['{:.0%}'.format(y) for y in y_vals])
    # x-axis formatting
    n = 10  # Keeps every nth label
    [l.set_visible(True) for (i, l) in enumerate(ax.xaxis.get_ticklabels()) if i % n != 0]
    for tick in ax.get_xticklabels():
        tick.set_rotation(45)
    # Other formatting
    plt.legend(labels=Freq.columns,loc='upper center', bbox_to_anchor=(0.5, 1.07),
              ncol=7)
    #plt.title('Percentage of each condition value from 2011 to 2020')
    plt.savefig("Percentage_condition_comparision.png", dpi=300)
    plt.show()
```
