```python
Freq=pd.concat([Master['Rating_2011'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2012'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2013'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2014'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2015'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2016'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2017'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2018'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2019'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame(),Master['Rating_2020'].value_counts(bins=[1,3,4,5,6,7,8,9]).to_frame()],axis=1)
Freq.reset_index(drop=True, inplace=True)
Freq.index=['<3','4','5','6','7','8','9']
Freq
```

```python
Freq=Freq.div(Freq.sum())
Freq=Freq.transpose()
Freq
```

```python
bottom_list = []

# Loops through each rating column. 
# Each column will take the sum of all columns before it as its base.
for i in Freq.columns:
    b = list(Freq.loc[:, Freq.columns[0]: i].sum(axis=1))
    bottom_list.append(b)
    
fig, ax = plt.subplots(figsize=(7,7))

x_ax = Freq.index

for i, v in enumerate(Freq.columns):
    if i == 0:
        # First column is at the bottom of the chart so bottom is 0 and doesn't need to be specified.
        plt.bar(x_ax, Freq[v],width=1)
    else:
        plt.bar(x_ax, Freq[v], bottom=bottom_list[i - 1],width=0.5)
# y-axis formatting
y_vals = ax.get_yticks()
ax.set_yticklabels(['{:.0%}'.format(y) for y in y_vals])
# x-axis formatting
n = 10  # Keeps every nth label
[l.set_visible(True) for (i, l) in enumerate(ax.xaxis.get_ticklabels()) if i % n != 0]
for tick in ax.get_xticklabels():
    tick.set_rotation(45)
# Other formatting
plt.legend(labels=Freq.columns,loc='upper center', bbox_to_anchor=(0.5, 1.07),
          ncol=7)
#plt.title('Percentage of each condition value from 2011 to 2020')
plt.savefig("Percentage_condition_comparision.png", dpi=300)
plt.show()
```
